```python
import matplotlib.pyplot as plt
```

```python
avg_scores=df.groupby('Access_to_Resources')['Exam_Score'].mean()
plt.figure()
plt.bar(avg_scores.index,avg_scores.values)
plt.xlabel("Acess to resources")
plt.ylabel("Average exam score")
plt.title("Impact of access to resources on exam score")
plt.show()
```

```python
avg_scores=df.groupby('Motivation_Level')['Exam_Score'].mean()
plt.figure()
plt.plot(avg_scores.index,avg_scores.values,marker='o')
plt.xlabel("Motivatio level")
plt.ylabel("Exam Score")
plt.title("Impact of Motivation level on Exam")
plt.show()
```

```python
plt.figure()
plt.boxplot(df['Exam_Score'])
plt.xlabel("Exam_Score")
plt.title("Box plot for exam score")
plt.show()
```

```python
plt.figure()
plt.boxplot(df['Hours_Studied'])
plt.xlabel("Hour Studied")
plt.title("Box plot for hours Studied")
plt.show()
```

```python
median_hours=df['Hours_Studied'].median()
df['Hours_Studied'].fillna(median_hours,inplace=True)
df['Hours_Studied'].isnull().sum()
```

```python
df['Teacher_Quality'].fillna(method='bfill',inplace=True)
df['Distance_from_Home'].fillna(method='ffill',inplace=True)
print("Teacher_Quality nulls:",df['Teacher_Quality'])
print("Distance_from_Home nulls:",df['Distance_from_Home'])
```

```python
df['Parental_Eduction_Level'].fillna('Undefined',inplace=True)
df['Parental_Eduction_Level'].isnull().sum()
```

```python
X_min=df['Hours_Studied'].min()
X_max=df['Hours_Studied'].max()
df['Hours_st_norm']=(df['Hours_Studied']-X_min)/(X_max-X_min)
df[['Hours_Studied','Hours_st_norm']].head()
```

```python
plt.figure()
plt.boxplot(df['Hours_st_norm'])
plt.ylabel("Normalization Hours Studied")
plt.title("Box plot of Hours_st_norm(Min-Max Normalization)")
plt.show()
```
